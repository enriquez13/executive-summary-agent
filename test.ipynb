```python
# verify
import os

api_key = os.getenv("GROQ_API_KEY")
print(f"Variable found? {' YES' if api_key else ' NO'}")
if api_key:
    print(f"Key (first 8 characters): {api_key[:8]}...")
    print(f"Total length: {len(api_key)} characters")
    
    # Prueba mínima de conexión
    from langchain_groq import ChatGroq
    try:
        llm = ChatGroq(groq_api_key=api_key, model_name="llama-3.3-70b-versatile")
        respuesta = llm.invoke("SUCCESSFUL CONNECTION")
        print(f"\n Test of LLM: {respuesta.content}")
    except Exception as e:
        print(f"\n Connection error: {type(e).__name__}: {e}")
else:
    print("\n You need to restart VS Code or the terminal.")
```
